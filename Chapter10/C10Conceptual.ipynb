### (c) Explain how this model can be thought of as an ordinary feedforward neural network with the individual pixels as inputs, and with constraints on the weights in the hidden units. What are the constraints?

* The input would be each pixel from the orginal image so 32 x 32 = 1024 inputs. The next hidden layer would be the convolution filter. Its weights would be the numbers inside the 5 x 5 convolution filter. The constraints would be it only takes from the current node to current node + 24. The nodes would have to be arrange in the correct order like taking several 5x5 convolutional filters of the image.

* The hidden node would then add up all 25 nodes and another constraint is applied. Either mod 255 since the pixel value can never go above that or some subtraction to always make it less than 256. Lastly, you could constrain the weights for these 25 muliplications so that even when every value is added up it is always below 256.

* Lastly the output layer would organize this into a 28 by 28 image. Since we have 3 filters we would get 3 images.

## Question 5 In Table 10.2 on page 426, we see that the ordering of the three methods with respect to mean absolute error is diferent from the ordering with respect to test set R2. How can this be?

* The mean absolute error takes the absolute difference between the prediction and the actual value. While $R^{2}$ denominator will stay the same for all 3 methods, its numerator will change based on $y$. It then squares this difference. Thus the estimates that are further away will contribute harsher to $R^{2}$ than mean absolute error, hence the difference.
