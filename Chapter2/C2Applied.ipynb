```python
Auto = Auto.drop(Auto[Auto.values == '?'].index)
```

```python
Auto.dtypes
```

```python
data = {
        'quant' : ['mpg', 'cylinders', 'horsepower' ,'weight', 'acceleration', 'year'],
        'qual' : ['origin' , 'name']
        }
Auto[data['quant']].max()
```

```python
for col in data['quant']:
  print('Min', col, Auto[col].min(), 'Max', col, Auto[col].max())

```

```python
Auto['horsepower'] = Auto.horsepower.astype(float)
```

```python
for col in data['quant']:
  print('Mean', col, Auto[col].mean(), 'Std', col, Auto[col].std())
```

```python
auto_quant = Auto[data['quant']]
pd.DataFrame({
              'min': auto_quant.min(),
              'max': auto_quant.max(),
              'mean': auto_quant.mean(),
              'std': auto_quant.mean()
              })
```

```python
auto_edit = auto_quant.drop(auto_quant.index[10:85])
pd.DataFrame({
              'min': auto_edit.min(),
              'max': auto_edit.max(),
              'mean': auto_edit.mean(),
              'std': auto_edit.mean()
              })

# 11 drop rows 10-85
```
