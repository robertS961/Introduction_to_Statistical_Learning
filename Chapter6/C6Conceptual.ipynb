## Question 6 - We will now explore (6.12) and (6.13) further.

###  (a) Consider $∑_{j=1}^{p}(y_{j} - B_{j})^{2} + λ∑_{j=1}^{p}B_{j}^{2}$ (6.12) with p = 1. For some choice of $y_{1}$ and $λ > 0 $, plot (6.12) as a function of $β_{1}$. Your plot should confrm that (6.12) is solved by (6.14).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def sixtwelve(y: int, b: list[int], lam: int) -> int:
  return (y - b)**2 + (lam * b**2)

lam, y = 5, 10
b = np.arange(-20, 20)
plt.plot(b, sixtwelve(y,b,lam))
plt.xlabel('b', fontsize = 14)
plt.ylabel('formula', fontsize = 14)
plt.title(' b vs formula', fontsize = 16)
Br = y/ (1 + lam)
plt.axvline(x = Br, color = 'r')
```

#### Results (a)

We can see that (6.14) does indeed prove the above forumula (6.12). We can see the min is indeed accurate by the formula.

(6.14)  $B_{j}^{R} = \frac{y}{1 + λ}$

### (b) Consider (6.13) with $p = 1$. For some choice of $y_{1}$ and $λ > 0 $, plot (6.13) as a function of $β_{1}$. Your plot should confirm that (6.13) is solved by (6.15).

(6.13):  $∑_{j=1}^{p}(y_{j} - B_{j})^{2} + λ∑_{j=1}^{p}|B_{j}|$


```python
def sxithirteen(y: int, b: list[int], lam: int) -> int:
  return (y-b)**2 +(lam* abs(b))

lam,y = 5, 10
b = np.arange(-40, 40)
plt.plot(b, sxithirteen(y,b,lam))
plt.xlabel('b', fontsize = 14)
plt.ylabel('formula', fontsize = 14)
plt.title(' b vs formula', fontsize = 16)

op1, op2, op3 = y - (lam/2), y + (lam/2), 0
if y > lam/2:
  plt.axvline(x= op1, color = 'r')
elif y< -(lam/2):
  plt.axvline(x = op2, color = 'r')
else:
  plt.axvline(x=op3, color = 'r')
```

#### Results (b)
